```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

font_axis = {
    "family": "calibri",
    "color": "black",
    "size": 12
}

font_title = {
    "family": "calibri",
    "color": "black",
    "size": 14
}

plt.title("Number of subsequences", fontdict=font_title)
plt.xlabel("Subsequences", fontdict=font_axis)
plt.ylabel("Quantity", fontdict=font_axis)

plt.bar(x, y, linestyle=":", linewidth=2, color="k")
plt.grid(color="k", linestyle=":", linewidth=1)
plt.show()
```
