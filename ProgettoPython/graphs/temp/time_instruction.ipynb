```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4])
y = np.array([3, 8, 1, 10])

font_axis = {
    "family": "calibri",
    "color": "black",
    "size": 12
}

font_title = {
    "family": "calibri",
    "color": "black",
    "size": 14
}

plt.title("Instruction time", fontdict=font_title)
plt.xlabel("Instruction", fontdict=font_axis)
plt.ylabel("Time", fontdict=font_axis)

plt.grid(color="k", linestyle=":", linewidth=1)
markerline, stemlines, baseline = plt.stem(x, y, linefmt="k:", markerfmt="o", basefmt="k", bottom=0)
plt.setp(stemlines, "linewidth", 3)
plt.show()
```
