```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 1, 2, 2, 3, 3, 4, 4])
y = np.array([1, 2, 2, 3, 3, 4, 4, 5])

font_axis = {
    "family": "calibri",
    "color": "black",
    "size": 12
}

font_title = {
    "family": "calibri",
    "color": "black",
    "size": 14
}

plt.title("Instruction time", fontdict=font_title)
plt.xlabel("Instruction", fontdict=font_axis)
plt.ylabel("Time", fontdict=font_axis)

plt.plot(x, y, marker="o", linestyle=":", linewidth=2, markeredgecolor="k", markerfacecolor="k", color="k", markersize=10)
plt.grid(color="k", linestyle=":", linewidth=1)
plt.show()
```
